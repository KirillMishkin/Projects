Данная структура должна быть одинаковая для вакансий с обоих сайтов. Общий результат можно вывести с помощью dataFrame через pandas.

### Импорт библиотек

```python
import requests
import pandas as pd 
import numpy as np
```

```python
def pars_hh(number_page,name_vacancies):
    main_link = 'https://api.hh.ru/vacancies'
    my_data = []
    for i in range(number_page):
        params = {'text':name_vacancies,
                  'page':i}
        response = requests.get(main_link, params = params )
        my_data.append(response.json())
    if number_page<=2:
        my_data[0]['items']+=my_data[1]['items']
    else:
        for i in range(1,number_page):
            my_data[0]['items']+=my_data[i]['items']
    return my_data[0]
```

#### Количество страниц для анализа

```python
number_page = int(input())
```

#### Наименование должности

```python
name_vacancies = input()
```

```python
data = pars_hh(number_page,name_vacancies)
```
